```python
import pandas as pd 
import numpy as np
weather_df = pd.read_csv('New York City weather data.csv')
```

```python
weather_df.head()
```

```python
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])
```

```python
weather_df.info()
```

```python
reviews_df = pd.read_csv('all_reviews_capstone.csv')
```

```python
reviews_df.head()
```

```python
reviews_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
reviews_df.info()
```

```python
reviews_df['review_date'] = pd.to_datetime(reviews_df['review_date'])
```

```python
reviews_df.info()
```

```python
final_df = pd.merge(reviews_df, weather_df, left_on = 'review_date', right_on = 'datetime')
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
cols = list(final_df.columns)
```

```python
cols
```

```python
final_df.drop(['user_name',
 'name',
 'datetime',
 'tempmax',
 'tempmin',
 'feelslikemax',
 'feelslikemin',
 'feelslike',
 'dew',
 'humidity',
 'precipprob',
 'precipcover',
 'preciptype',
 'windgust',
 'winddir',
 'sealevelpressure',
 'cloudcover',
 'visibility',
 'solarradiation',
 'solarenergy',
 'uvindex',
 'severerisk',
 'sunrise',
 'sunset',
 'moonphase',
 'description',
 'stations', 
 'snowdepth'], axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df['state'] = final_df['home_town'].str[-2:]
final_df['city'] = final_df['home_town'].str[:-4]
```

```python
final_df.head(3)
```

```python
final_df['review_date'] = final_df['review_date'].astype(str)
```

```python
final_df[['year', 'month', 'day']] = final_df['review_date'].str.split('-', expand = True)
```

```python
final_df.head(1)
```

```python
final_df.dropna(axis=0, inplace = True)
```

```python
final_df.to_csv('Capstone_model_building_dataset.csv')
```
